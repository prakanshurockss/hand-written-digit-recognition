# Images Collection

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
#read the dataset
df=pd.read_csv("train.csv")
df
```

# 1

```python
df1=df[df['label']==1]
df1.reset_index(drop=True, inplace=True)
df1=df1.drop('label',axis=1)
df1
```

```python
df1.shape
```

```python
a=''
for i in range(200):
    array1=df1.loc[i].to_numpy()
    array1 = np.reshape(array1, (28, 28))
    plt.imshow(array1,cmap=plt.cm.binary)
    plt.axis("off")
    val=("%s%s" % (a, i))
    plt.savefig(('{}/'+val).format('captured_images/1'))


```
